```python
# INSERT California_Gasoline_Prices.csv (Eric)
California_Gasoline_Prices = pd.read_csv('https://raw.githubusercontent.com/EricChen0803/DATASET-TEMP/main/Datasets/California_Gasoline_Prices.csv')
# Using apply function to change the month data to year
def keepYear(s):
    return int(s.split('-')[1])
California_Gasoline_Prices['Year'] = California_Gasoline_Prices.get('Month').apply(keepYear)
# Only keeping the data from 2020 to 2023
California_Gasoline_Prices = California_Gasoline_Prices[California_Gasoline_Prices.get("Year")>=20]
California_Gasoline_Prices = California_Gasoline_Prices[California_Gasoline_Prices.get("Year") <=23]
California_Gasoline_Prices['Year'] = California_Gasoline_Prices.get('Year')+2000
# Grouping to get the price per gallon for each year
California_Gasoline_Prices = California_Gasoline_Prices.get(['Year','California All Grades All Formulations Retail Gasoline Prices Dollars per Gallon']).groupby("Year").mean()
California_Gasoline_Prices
```
